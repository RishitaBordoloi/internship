# Data Processing

```python
dataset=pd.get_dummies(df,columns=['sex','cp','fbs','restecg','exang','slope','ca','thal'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
```

```python
columns_to_scale=['age','trestbps','chol','thalach','oldpeak']
dataset[columns_to_scale]=standardScaler.fit_transform(df[columns_to_scale])
```

```python
dataset.head()
```

```python
y = dataset['target']
X = dataset.drop(['target'], axis=1)
```

# K Nearest Neighbr

```python
from sklearn.model_selection import cross_val_score
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors=k)
    score=cross_val_score(knn_classifier,X,y,cv=10)
    knn_scores.append(score.mean())
```

```python
plt.rcParams['figure.figsize'] = (20,8)
plt.plot([k for k in range(1,21)],knn_scores, color='red')
for i in range(1,21):
    plt.text(i,knn_scores[i-1],(i,knn_scores[i-1]))
plt.xticks([i for i in range(1,21)])
plt.xlabel('Number of Neighbrs(k)')
plt.ylabel('Scores')
plt.title(' Neighbors Classifier scores for different k values')
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=12)
score=cross_val_score(knn_classifier,X,y,cv=10)
```

```python
q=score.mean()
q
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randomforest_classifier=RandomForestClassifier(n_estimators=15)
score1=cross_val_score(randomforest_classifier,X,y,cv=10)
```

```python
w=score1.mean()
w
```

# Decission Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion = 'entropy',)
score2=cross_val_score(model,X,y,cv=10)
```

```python
e=score2.mean() 
e
```

# Support vector machine

```python
from sklearn import svm
def run_svm(c):
    svm_model = svm.SVC(kernel='linear', C=c, probability=True)
    svm_model.fit(X, y)
    svm_score = svm_model.score(X,y)
    df = pd.DataFrame([[c, svm_score]], columns=['c_parameter', 'score'])
    return df
```

```python
output1 = run_svm(10)
```

```python
output2 = run_svm(100)
```

```python
output3 = run_svm(1000)
```
